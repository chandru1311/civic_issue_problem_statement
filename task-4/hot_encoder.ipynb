```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

df = pd.read_csv("civic_issue_dataset.csv")
```

```python
data = df.copy()

categorical_cols = ["Issue_Type", "Department", "Zone"]

dummies = pd.get_dummies(data[categorical_cols], dtype=int)
merged = pd.concat([data.drop(columns=categorical_cols), dummies], axis="columns")

X = merged.drop(columns=["Report_ID", "Date_Submitted", "Severity_Label", "Priority_Score"])
y = merged["Priority_Score"]

model = LinearRegression()
model.fit(X, y)

print(model.score(X, y))
```

```python
dfle = df.copy()

le = LabelEncoder()
for col in dfle.select_dtypes(include="object").columns:
    if col not in ["Report_ID", "Date_Submitted", "Severity_Label"]:
        dfle[col] = le.fit_transform(dfle[col])

y = dfle["Priority_Score"].values

ct = ColumnTransformer(
    [("zone", OneHotEncoder(drop="first", sparse_output=False), [0])],
    remainder="passthrough"
)

X = ct.fit_transform(X)

model.fit(X, y)

X_pred = ct.transform(X[:1])
print(model.predict(X_pred))
```
