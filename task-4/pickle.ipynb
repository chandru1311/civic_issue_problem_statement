```python
import pandas as pd
import numpy as np
from sklearn import linear_model

df = pd.read_csv("civic_issue_dataset.csv")
df.head()
```

```python
X = df[["Severity_Score", "Num_Upvotes", "Has_Photo", "Population_Density"]]
y = df["Priority_Score"]

model = linear_model.LinearRegression()
model.fit(X, y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

sample = pd.DataFrame({
    "Severity_Score": [3],
    "Num_Upvotes": [5],
    "Has_Photo": [1],
    "Population_Density": [2000]
})
print("Prediction for sample input:", model.predict(sample))
print("R² Score:", model.score(X, y))
```

```python
import pickle

with open('model_pickle.pkl', 'wb') as file:
    pickle.dump(model, file)

with open('model_pickle.pkl', 'rb') as file:
    mp = pickle.load(file)

print("Loaded Coefficients:", mp.coef_)
print("Loaded Intercept:", mp.intercept_)
print("Loaded Prediction for sample input:", mp.predict(pd.DataFrame({
    "Severity_Score": [3],
    "Num_Upvotes": [5],
    "Has_Photo": [1],
    "Population_Density": [2000]
})))
```

```python
import joblib

joblib.dump(model, 'model_joblib.joblib')
mj = joblib.load('model_joblib.joblib')

print("Loaded Coefficients:", mj.coef_)
print("Loaded Intercept:", mj.intercept_)
print("Loaded Prediction for sample input:", mj.predict(pd.DataFrame({
    "Severity_Score": [3],
    "Num_Upvotes": [5],
    "Has_Photo": [1],
    "Population_Density": [2000]
})))
```
